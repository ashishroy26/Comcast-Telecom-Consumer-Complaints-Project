```python
data["Status"].unique()    #status types
```

```python
#mentioning "Open" and "Pending" status as "Open" and "Solved" status as "Closed"
```

```python
data["New status"]=data["Status"].map({"Closed":"Closed", "Open":"Open","Solved":"Closed","Pending":"Open"})
```

```python
data[["State","New status"]].head()
```

```python
#Provide state wise status of complaints in a stacked bar chart. Use the categorized variable from Q3. Provide insights on:
#Which state has the maximum complaints
#Which state has the highest percentage of unresolved complaints
#- Provide the percentage of complaints resolved till date, which were received through the Internet and customer care calls
```

```python
datanew=data[["State","New status"]].groupby(["State","New status"]).size().unstack().reset_index()
```

```python
datanew.fillna(0,inplace=True)
```

```python
datanew.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
fig= plt.figure(figsize=(20,10))

plt.bar(datanew['State'], datanew['Closed'], label='Closed')
plt.bar(datanew['State'], datanew['Open'], label='Open')

plt.xticks(rotation =90, size=14)
plt.legend(loc="upper right",bbox_to_anchor=(0.9,0.9))
plt.xlabel('States',size=20)
plt.ylabel('Number of Complaints',size=20)
plt.show()
```

```python
datanew.fillna(0, inplace=True)
sum=datanew["Closed"]+datanew["Open"]
datanew["% of Closed"]=((datanew["Closed"]/sum)*100).round(2)
datanew["% of Open"]=((datanew["Open"]/sum)*100).round(2)
```
